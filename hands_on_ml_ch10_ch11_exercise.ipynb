* 계단함수(헤비사이드 계단함수)
> $heavyside(z) \begin{cases}
0, & \mbox{if }n\mbox{ z <0} \\
1, & \mbox{if }n\mbox{ z $\leqq$0}
\end{cases} $

* 로지스틱함수(시그모이드)
> $\sigma(z)=1/(1+exp(-z))$. 
> 출력범위 0~1. 양극단에서 기울기가 급격히 감소, 오차 그레이언트를 역전파 잘못함


* 하이퍼볼릭 탄젠트 함수(쌍곡 탄젠트 함수)
> $tanh(z) = 2\sigma(2z)-1$. 로지스틱 함수처럼 S자모양. 연속&미분가능 출력범위가 -1~1. 훈련초기에 출력을 원점 근처로 모으는 경향이 있고 빠른 수렴이 되도록 기여하는 부분이 있음(양극단에서 기울기가 급격히 감소, 오차 그레이언트를 역전파 잘못함)


* ReLU 함수
> $ReLU(z) = max(0,z)$ 연속적이지만 z=0에서 미분불가.z<0에서 도함수는 0.
> 계산속도가 빠르다는 장점있어서 기본활성화 함수.  출력에 최대값이 없다는 점이 경사하강법의 문제 완화에 기여(Ch 11장에 등장)
